# Generate Data

```python
import numpy as np

N = 1000  # Number of data points

# Generate time values between 0 and 1000 as integers
time = np.round(np.linspace(0, 1000, N), decimals=0).astype(int)

# Define a function to introduce time dependency
def time_dependency(t):
    return np.sin(t * 2 * np.pi / 100)

# Generate random data points between 0 and 1
data = np.random.random(N)

# Scale the data to the desired range
data = data * 10

# Center the data around zero
data = data - np.mean(data)

# Add time dependency to the data
data = data + time_dependency(time)

# Round the data to 2 decimal places
data = np.round(data, 2)

# Print the generated time-series data
for t, d in zip(time[0:10], data[0:10]):
    print(f"Time: {t}, Data: {d}")
```

# Visualize the data

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(15, 5))

# Plot the time-series data
plt.plot(time, data)
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Random Time-Series Data')
plt.grid(True)
plt.show()
```

```python
# Set the range of data to visualize
start_index = 100
end_index = 200
time_range = time[start_index:end_index]
data_range = data[start_index:end_index]

# Set the figure size
plt.figure(figsize=(15, 5))
# Plot the time-series data within the specified range
plt.plot(time_range, data_range)
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Random Time-Series Data (Range: 100 to 200)')
plt.grid(True)
plt.show()
```
